```python
import numpy as np
```

```python
np.random.seed(1)
# function to select random values
```

```python
x=np.random.randint(0,50,100)
# 10 random integers between 0 and 50
```

```python
x
```

```python
y = x + np.random.normal(0,10,100)
# postive corelation with same noise
```

```python
y
```

```python
np.corrcoef(x,y)
```

```python
import matplotlib 
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')

plt.scatter(x,y)
plt.show()
```

```python
y = 100-x +np.random.normal(0,10,100)
#negative corelation with same noise
```

```python
y
```

```python
np.corrcoef(x,y)
```

```python
import matplotlib 
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')

plt.scatter(x,y)
plt.show()
```

```python
import pandas as pd
```

```python
df=pd.DataFrame({'a': np.random.randint(0,50,1000)})
df['b']=df['a']+np.random.normal(0,10,1000)
df['c']=100-df['a']+np.random.normal(0,10,1000)
df['d']=np.random.normal(0,10,1000)

df.corr()
```

```python
pd.plotting.scatter_matrix(df,figsize=(6,6))
plt.show()
```

```python
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)),df.columns)
plt.yticks(range(len(df.columns)),df.columns)
plt.colorbar()
plt.show()
```
